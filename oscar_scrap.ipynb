```python

s = requests.get(url="https://pt.wikipedia.org/wiki/%C3%93scar").text
soup = BeautifulSoup(s,"lxml")
```

```python
i = 0
for table in soup.find_all("table", { "class" : "wikitable sortable" }):
    head = []
    for name in table.findAll('th'):
        head.append(name.text)
    print(head)
    df = pd.DataFrame(columns=head)
    for row in table.find_all('tr'):
        cells = row.findAll('td')

        if len(cells) > 0:
            lista = []
            for r in cells:
                lista.append(r.text)
            df = df.append(pd.Series(lista,index= head),ignore_index=True)
    df.to_csv(path_or_buf= 'data_' + str(i) + '.csv') 
    print(df)
    i = i+1
```
