```python
from bs4 import BeautifulSoup
import urllib
import pandas as pd
import numpy as np
```

```python
import urllib.request
with urllib.request.urlopen("https://pt.wikipedia.org/wiki/%C3%93scar") as url:
    s = url.read()
#I'm guessing this would output the html source code?
soup = BeautifulSoup(s,"lxml")
```

```python
f = open("oscar.html",'w')
f.write(soup.prettify() )
```

```python
i = 0
for table in soup.find_all("table", { "class" : "wikitable sortable" }):
    head = []
    for name in table.findAll('th'):
        head.append(name.text)
    print(head)
    df = pd.DataFrame(columns=head)
    for row in table.find_all('tr'):
        cells = row.findAll('td')

        if len(cells) > 0:
            lista = []
            for r in cells:
                lista.append(r.text)
            df = df.append(pd.Series(lista,index= head),ignore_index=True)
    df.to_csv(path_or_buf= 'data_' + str(i) + '.csv') 
    show(df)
    i = i+1
```
